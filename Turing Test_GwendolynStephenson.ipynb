```python
''' 
THE TURING TEST

The chatbot responds to specific keywords or phrases defined in the reponses
dictionary and provides predefined reponses for these keywords.

if the user's input does not match any keywords, it returns a default response.

the conversation continues until the user enters "exit" to end it.

This is more of a "intelligent machine" where the responses correspond to the 
users question.

in a hypothetical scenario, the turing judge would NOT know this is a machine
'''

from datetime import datetime

#dictionary (reponses) contains a set of key-value pairs, chotbot repsonds
#based on keyword detected.
responses = {
    "hello": "Hello! How can I assist you today?",
    "how are you": "I'm just a computer program, so I dont have feelings, but I'm here to help",
    "bye": "Goodbye! feel free to return if you have more questions.",
    "thanks": "You're welcome! Is there anything else you'd like to know?",
    "old": "I don't have an age. I'm just a program running on a computer",
    "time": "The current date and time is: {current_date}",
}


''' 
With this, I put the generate reponse function above the chat funcation, and then I incporoated the ai_reponse within the while loop
so that I can obtain the results I wanted. The program works as intended. 
'''
#takes user's input as a paramter and determines the appropriate response.
def generate_response(user_input):
    for keyword, response in responses.items():
        if keyword in user_input:
            if keyword == "time":
                current_date = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                return response.format(current_date=current_date)
            else:
                return response
            
    #If no keyword is found, generate a default repsonse.
    return "I'm not sure I understand. Can you please rephrase your question?"
            

#chatbot interface, handles the conversation with the user
def chat():
    print("Part #2: - Turing simulation Chatbot")
    print("Where you can have pseudo conversation with an AI. Type 'exit' to end the conversation")

    while True:
        user_input = input("You: ")
        lower_input = user_input.lower()
        ai_response = generate_response(lower_input)

        if user_input == "exit":
            print("AI: Goodbye!")
            break
        else:
            print(f"You: {user_input}")
            print(f"AI: {ai_response}")

if __name__ == "__main__":
    chat()
```
